## find the odd numbers from a given list using a filter.list=[1,2,3,4,5,6,7,8,9]

```python
import numpy as n 
da=[1,2,3,4,5,6,7,8,9]
d=n.array(da)
print(list(filter(lambda num:num%2!=0,d)))
```

## find a list of squares of first five odd numbers using lambda and map function

```python
import numpy as n
l=[1,3,5,7,9]
d=n.array(l)
print(list(map(lambda num:num**2,d)))
```

```python
import pymysql 

# Database connection details
host = "localhost"
user = "nandu"
password = "yo"
database = "company1"  # Replace with your actual database name
table = "Employee"  # Replace with your actual table name

try:
    # Connect to the database
    connection = pymysql.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )

    with connection.cursor() as cursor:
        # Execute a SQL query
        sql = f"SELECT * FROM {table}"  # Adjust the query as needed
        cursor.execute(sql)

        # Fetch all rows from the executed query
        rows = cursor.fetchall()

        # Print the column names
        column_names = [desc[0] for desc in cursor.description]
        print("\t".join(column_names))

        # Print each row
        for row in rows:
            print("\t".join(str(value) for value in row))

except pymysql.MySQLError as e:
    print(f"Error: {e}")
finally:
    if connection:
        connection.close()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Load the dataset
t = pd.read_csv('D:\\fruit data.csv')

# Define colors for each fruit
fruit_colors = {
    'apple': 'r', 
    'orange': 'k', 
    'mandarin': 'm', 
    'lemon': 'y'
}

# Create a scatter plot for fruit width and height
for fruit_name, color in fruit_colors.items():
    # Use fruit_name to filter the DataFrame
    fw = t[t['fruit_name'] == fruit_name]['width']
    fh = t[t['fruit_name'] == fruit_name]['height']
    plt.scatter(fw, fh, color=color, label=fruit_name)  # Use fruit_name for labeling

plt.legend()
plt.xlabel('Fruit Width')
plt.ylabel('Fruit Height')
plt.title('Fruit Width and Height for Fruits')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
t=pd.read_csv('D:\\fruit data.csv')
plt.figure(figsize=(6,7))
color=['red','orange','magenta','yellow']
t['fruit_name'].value_counts().plot(kind='bar', color=color)
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.title('Count of each fruit')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
d=pd.read_csv("D:\\mtcars.csv")
#Histogram
mpg=d['mpg']
plt.hist(mpg,bins='auto',color='k',edgecolor='cyan')
plt.xlabel('Miles per gallon(mpg)'); plt.ylabel('Frequency')
plt.title('Highest frequency distribution')
plt.show()
#Scatter plot
wt=d['wt']
iv=range(len(d))
plt.scatter(iv,mpg,color='r',label='mpg')
plt.scatter(iv,wt,color='g',label='wt')
plt.title("Relationship between mpg and weight")
plt.legend()
plt.show()
#Bar plot
am=d['am'].value_counts()
color=['g','c']
plt.bar(am.index,am.values,color=color,width=0.2)
plt.xticks([0,1],('0-Automatic','1-Manual'))
plt.xlabel('Transmission type of cars');plt.ylabel('No. of cars')
plt.title('Frequency distribution of transmission type of cars')
plt.show()
#Box plot
sn.boxplot(mpg,color='lightblue')
plt.xlabel('MPG');plt.ylabel('Values')
plt.title('Box plot of mpg values')
plt.show()
```
